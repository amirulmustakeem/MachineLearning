```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import neighbors,metrics
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
#load datasets using pandas
data = pd.read_csv("car.data", names=["buying","maint","doors","persons","lug_boot","safety","class"])

data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
X= data[[
    "buying",
    "maint",
    "safety"
]].values
y= data[["class"]]
```

```python
#convert data string into numeric

#X
Le=LabelEncoder()
for i in range(len(X[0])):
    X[:,i]=Le.fit_transform(X[:,i])
```

```python
#y
label_maping={
    'unacc':0,
    'acc':1,
    'good':2,
    'vgood':3
}
y['class']= y['class'].map(label_maping)
y=np.array(y)
```

```python
y
```

```python
#create model KNN
knn=neighbors.KNeighborsClassifier(n_neighbors=25, weights='uniform')
#split into train and test dataset with 20% test data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
knn.fit(X_train,y_train)
```

```python
prediction = knn.predict(X_test)
accuracy = metrics.accuracy_score(y_test, prediction)
print("prediction",prediction)
print("accuracy:",accuracy) 
```

```python
for i in range(len(X)):
    print("actual values:",y[i])
    print("predicted values:",knn.predict(X)[i])
```
