```python
import numpy as np
import matplotlib.pyplot as plt
import mglearn
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
%matplotlib inline
```

```python
dataset= load_iris()
dataset.keys()
```

```python
print(dataset["data"][:5])
```

```python
print(dataset["target"])
    
```

```python
print(dataset["target_names"])
```

```python
X_train,X_test,y_train,y_test = train_test_split(dataset["data"],dataset["target"],test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
iris_dataframe= pd.DataFrame(X_train, columns=dataset.feature_names)
```

```python
grr=pd.plotting.scatter_matrix(iris_dataframe,c=y_train,figsize=(15,15),marker='o',diagonal='kde',s=60,alpha=.8,cmap=mglearn.cm3)
```

```python
#knn model
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
prediction = knn.predict(X_test)
```

```python
prediction
```

```python
print("accuracy:",knn.score(X_test,y_test))
```

```python
x_new = np.array([[5,2.9,1,0.2]])
x_new.shape
```

```python
test_prediction = knn.predict(x_new)
print("Predicted target name: {}".format(dataset["target_names"][test_prediction]))
```
